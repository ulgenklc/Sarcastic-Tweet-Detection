```python
import pandas as pd
import numpy as np
```

### Get the data

```python
path = '/Users/bengieru/Desktop/NMA-DL/iSarcasm/iSarcasm_dataset/'

test_raw = 'rawdata/test.txt'
train_raw = 'rawdata/train.txt'

test_ids_raw = 'rawdata/test_only_ids.txt'
train_ids_raw = 'rawdata/train_only_ids.txt'

test_meta = 'metadata/metadata_test.csv'
train_meta = 'metadata/metadata_train.csv'

path_test_raw = path + test_raw
path_train_raw = path + train_raw

path_test_ids_raw = path + test_ids_raw
path_train_ids_raw = path + train_ids_raw

path_test_meta = path + test_meta
path_train_meta = path + train_meta

test_rawdata = pd.read_csv(path_test_raw, delimiter = '\t')
train_rawdata = pd.read_csv(path_train_raw, delimiter = '\t')

test_ids_rawdata = pd.read_csv(path_test_ids_raw, delimiter = '\t')
train_ids_rawdata = pd.read_csv(path_train_ids_raw, delimiter = '\t')

test_metadata = pd.read_csv(path_test_meta)
train_metadata = pd.read_csv(path_train_meta)
```

## Adjust the annoying zeros at the end of tweet ids due to the excel

## Create empty DataFrames of the respective sizes

```python
a = np.empty((3275,3))
a[:] = np.NaN
b = np.empty((809,3))
b[:] = np.NaN
train_cleandata = pd.DataFrame(a, columns = ['text', 'label', 'source'])
test_cleandata = pd.DataFrame(b, columns = ['text', 'label', 'source'])
```

### Adjust the tweet ids that doesn't exist anymore and label tweets as Sarcastic(True) and Not_Sarcastic(False)

```python
for i,e in enumerate(train_rawdata['tweet_id']):
    for j,f in enumerate(train_metadata['id']):
        if e == f:
            train_cleandata['text'].iloc[j] = train_metadata['text'].iloc[j]
            train_cleandata['source'].iloc[j] = int(3)
            if train_rawdata['sarcasm_label'].iloc[i] == 'sarcastic':
                train_cleandata['label'].iloc[j] = True
            elif train_rawdata['sarcasm_label'].iloc[i] == 'not_sarcastic':
                train_cleandata['label'].iloc[j] = False
            #train_cleandata['text'][i] = train_metadata['text'][j]

for i,e in enumerate(test_rawdata['tweet_id']):
    for j,f in enumerate(test_metadata['id']):
        if e == f:
            test_cleandata['text'].iloc[j] = test_metadata['text'].iloc[j]
            test_cleandata['source'].iloc[j] = int(3)
            if test_rawdata['sarcasm_label'].iloc[i] == 'sarcastic':
                test_cleandata['label'].iloc[j] = True
            elif test_rawdata['sarcasm_label'].iloc[i] == 'not_sarcastic':
                test_cleandata['label'].iloc[j] = False
            #train_cleandata['text'][i] = train_metadata['text'][j]
```
