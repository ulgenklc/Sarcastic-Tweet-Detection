```python
import pandas as pd
import numpy as np
```

```python
path = '/Users/bengieru/Desktop/NMA-DL/iSarcasm/iSarcasm_dataset/'

test_raw = 'rawdata/test.txt'
train_raw = 'rawdata/train.txt'

test_meta = 'metadata/tweets_test.csv'
train_meta = 'metadata/tweets_train.csv'

path_test_raw = path + test_raw
path_train_raw = path + train_raw

path_test_meta = path + test_meta
path_train_meta = path + train_meta

test_rawdata = pd.read_csv(path_test_raw, delimiter = '\t')
train_rawdata = pd.read_csv(path_train_raw, delimiter = '\t')

test_metadata = pd.read_csv(path_test_meta)
train_metadata = pd.read_csv(path_train_meta)
```

```python
columns_metadata = train_metadata.columns.tolist()
columns_rawdata = train_rawdata.columns.tolist()
columns_metadata, columns_rawdata
```

```python
train_cleandata = train_rawdata
test_cleandata = test_rawdata
test_cleandata['text'] = np.NaN
train_cleandata['text'] = np.NaN
```

```python
for i,e in enumerate(train_rawdata['tweet_id']):
    for j,f in enumerate(train_metadata['id']):
        if e == f:
            train_cleandata['text'].iloc[i] = train_metadata['text'].iloc[j]
            #train_cleandata['text'][i] = train_metadata['text'][j]

for i,e in enumerate(test_rawdata['tweet_id']):
    for j,f in enumerate(test_metadata['id']):
        if e == f:
            test_cleandata['text'].iloc[i] = test_metadata['text'].iloc[j]
            #test_cleandata['text'][i] = test_metadata['text'][j]
```

```python
#train_cleandata.to_csv(path + 'cleandata/train.csv')
train_cleandata.head()
```

```python
#train_cleandata.to_csv(path + 'cleandata/test.csv')
test_cleandata.head()
```

```python
for i,e in enumerate(train_cleandata['text']):
    if e == np.NaN:
        train_cleandata.drop(train_cleandata.index[i], axis = 0)
        
for i,e in enumerate(test_cleandata['text']):
    if e == np.NaN:
        test_cleandata.drop(test_cleandata.index[i], axis = 0)
```

```python
train_cleandata.to_csv(path + 'cleandata/train.csv')
#train_cleandata.dropna()
train_cleandata.head()
```
